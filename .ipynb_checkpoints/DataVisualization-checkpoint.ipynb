### Null Values

There are two main methods to dealing with null values. 

1) The simplest is just to remove them all together (either the row if one particular entry is missing a majority of the values or the column if the majority of that column is null), but that's often not the best method since it does still throw away some valid data. There are two ways to drop null data, by column (any variable that is above some chosen threshold of null %) or by row (a row missing ANY of the variables kept after feature engineering should be thrown away)

2) Impute the missing data from the remaining data. If there is sufficient data then this is often the preferred method, since if done correctly it will retain the shape of the data (distribution) while not throwing away those values that WERE measured. There are a few ways to do this and it highly depends on what the distribution looks like and how many values are missing. Note that in that case where an entry or column is overwhelmingly null (>95%) interpolating the missing data from the very small fraction of non-null WILL generate a large amount of bias in the data, no matter what you use. You can get away with maybe up to 80-90% null data with algorithms like Non-Negative Matrix Facorization (https://www.biorxiv.org/content/10.1101/647560v2.full) to try and infer the missing data. For more reasonable percentages, some popular ones are:

* A constant value that has meaning within the domain, such as 0, distinct from all other values.
* A value from another randomly selected record.
* A mean, median or mode value for the column.
* A value estimated by another predictive model.

**Important Note:** Any imputing done to the training data must be done to all future data as well in order for the model to be used on that data

It's worth noting that not all models are sensitive to missing values. Many tree based algorithms treat null values as essentially just another unique value for that feature). Xgboost also handles null values and works very well with structured data (which this is). Due to the overwhelming number of null values for some of these features, if they are noticeably linked to sepsis prediction during our analysis and decided to be kept, mean imputation (the default) is a bad idea.

### What do these null values mean?

An important thing to distinguish in any data science problem is what null represents. In this case, why are some of these values null and others aren't. For a proper understanding, we would likely need to understand how the measurement process of each of these works.

**Potential symptoms of sepsis:**
* Change in mental status
* A first (upper) number in a blood pressure reading — also called the systolic pressure — that's less than or equal to 100 millimeters of mercury (mm Hg)
* Respiratory rate higher than or equal to 22 breaths a minute

These two make up the quick Sequential Organ Failure Assessment (qSOFA). Essentially a 'quick' assessment that tells the physician if further steps have to be taken, since the full SOFA score requires laboratory tests.

If sepsis progresses into sepsis shock and causes blood pressure to drop, it can lead to death. Potential symptoms of the septic shock phase are:
* The need for medication to maintain blood pressure greater than or equal to 65 millimeters of mercury (mm Hg).
* High levels of lactic acid in your blood (serum lactate) after you have received adequate fluid replacement. Having too much lactic acid in your blood means that your cells aren't using oxygen properly.

People in ICU are at risk of sepsis due to the risk of infections. So the data we have received is simply the monitoring data hospitals provide for patients in ICUs while they are recovering. And since all of the overwhelmingly sparse variables come from labratory results, the reasonable assumption is that most ICU patients don't have lab work done.

In this case, the best option would have been to talk to physician that understands the methodology and procedure for patients suspected of sepsis in the ICU. But short of that, we can make some reasonable assumptions from the information given

1) Certain lab tests are only done after being requested due to certain triggers (qSOFA being one of them, but papers suggest there are a number of clinical signs and symptoms)

2) The request of lab tests (due to costs) only happens if physicians or caretakers have reason to assume something is going wrong

### Feature Selection

When a feature is not important to the model, including it can waste time at best and cause incorrect fitting at worst. Also, especially in a case where many of the values are missing (like with Troponin, which is 99.9% null values), if the feature is not particularly useful throwing it away can help reduce bias.

Also due to the nature of sparse test results, we have another problem, overfitting. For example, pretend Troponin has nothing to do with sepsis, however only patients at high risk of sepsis get the test taken where Troponin (along with others) is measured, therefore it is returned with the other results. Now there shows a high correlation between troponin and sepsis during statistical analysis when there really isn't (really, this is the part where you should just go consult a physician with medical knowledge of the diagnosis procedure)

### Understanding sepsis diagnosis

*However*, as mentioned when discussing null values, in this case a large number of those values are due to lab tests being expensive and sporadic. But consequently, it means they aren't ordered unless there is reason to believe the patient has an issue (not necessarily sepsis). While there are generic statistic and filter based feature selection methods, which will still be considered, in this case it is most likely more useful to base feature selection in part on external methodology, namely those used to determine if a patient is at risk of sepsis.

Two things that should be noted:

1) Lab tests will return all variables requested of that time (ie: requesting bloodworks gives you all the information, there is no 'sepsis only' bloodworks test) so there is still the possibility of eliminating some of the variables with high null percentages

2) There is no sepsis specific test, many of the following methods are use to determine the degree of illness in a patient and if they require critical care.

To elaborate on the second point, unlike positive/negative measurements sepsis is always a guessing game. Many of the methods like SOFA or NEWS are a scoring system where when a patient has more than X points they are at risk and recommended critical care, additional tests, treatments, etc. The treatment for sepsis overlap with the treatment for many other critical infections or complications. So essentially, if a patient scores high enough on SOFA, NEWS, SIRS, or whichever method the hospital is using to determine critical care requirements, they should also be marked for sepsis.

Ideally we'd like to know which method the hospital where the ICU is was using since that would determine what variables they were looking at. In lieu of that, we'll consider the most popular methods currently: SOFA, NEWS and SIRS. SOFA and NEWS are the more favored methods according to several sites and papers so we'll consider those.

*Update*: While the KP website doesn't specify, it turns out the PhysioNet challenge it got the data from uses the Sepsis-3 definition of sepsis (https://litfl.com/sepsis-definitions-and-diagnosis/) so we're going with SOFA (@future me it might be worth it to go back and use the NEWS definition to see if there's any changes)

**SOFA variables** (taken from MDCALC: https://www.mdcalc.com/sequential-organ-failure-assessment-sofa-score)
* PaO2 (partial pressure of oxygen)
* FiO2 (fraction of inspired oxygen)
* Glasgow Coma Scale
* Platelets 
* MAP (mean arterial pressure)
* Bilirubin
* Creatinine

Of these, the ones that we have in our data are
* FiO2
* Platelets
* MAP
* Bilirubin
* Creatine

However some quick research shows that SaO2 is directly related to PaO2 so to make up for not having the direct measurement, we can include that as well. That leaves the only missing value as the Glasgow Coma Scale, which doesn't really have any replacements.

And while it's not used in SOFA (most likely because sepsis and septic shock definitions are split up), the definition of septic shock includes lactate.

So let our starting features be all of these along with variables with high ovservation rates (vital signs and demographics). This gives us the following:
<b>
1. HR 	Heart rate (beats per minute)
2. O2Sat 	Pulse oximetry (%)
3. Temp	Temperature (Deg C)
4. SBP	Systolic BP (mm Hg)
5. MAP	Mean arterial pressure (mm Hg)
6. DBP	Diastolic BP (mm Hg)
7. Resp	Respiration rate (breaths per minute)
8. FiO2	Fraction of inspired oxygen (%)
9. SaO2	Oxygen saturation from arterial blood (%)
10. Lactate 	Lactic acid (mg/dL)
11. Bilirubin_total		Total bilirubin (mg/dL)
12. Platelets (count * 10^3/µL)
13. Creatinine (mg/dL)
14. Age		Years (100 for patients 90 or above)
15. Gender	Female (0) or Male (1)
16. HospAdmTime		Hours between hospital admit and ICU admit
17. ICULOS	ICU length-of-stay (hours since ICU admit)
</b>

### Value Imputation
While imputing with the mean or median makes no sense in this situation, forward filling and back filling are potentially valid. Part of the reason test results are so sparse is because tests are not ordered often, but also because they are only recorded in the hour that the results return. In other words, if a patient is in the ICU for 10 more hours after their test results, those 10 hours would be missing result values even though whether or not they have sepsis would be determined using those previous results. 

#### Forward filling
We want to only fill in the values of the given patient and not spill over into the next patient (the original data had a file per patient, but we have one csv with all the patients).

#### Backward filling
For models that cannot handle null values, we need to fill all missing values. We can use backward filling to fill in previous results as well since any decision made on that patient using test results will simply propogate backwards. 

```python
#temp: in the interest of time when just testing, use the smaller set for now
train_small = pd.read_csv("./Data/train_small.csv")
train_small.drop('pH', axis=1, inplace=True)  
missing_column = ['Age','Gender','Unit1','Unit2','HospAdmTime','ICULOS','ptn','sepsis']

fillcol = ['HR','O2Sat','Temp','SBP','MAP','DBP','Resp','BaseExcess','HCO3','FiO2','PaCO2','SaO2','AST',
               'BUN','Alkalinephos','Calcium','Chloride','Creatinine','Bilirubin_direct','Glucose','Lactate',
               'Magnesium', 'Phosphate','Potassium','Bilirubin_total','TroponinI','Hct','Hgb','PTT','WBC',
               'Fibrinogen','Platelets']

c = train_small[missing_column]
ca = train_data_raw[missing_column]
```

```python
#groupby ptn first then fill-forward/backward-fill the vital and test values only
train_small_fill = train_small.groupby('ptn')[fillcol].transform(lambda v: v.ffill().bfill())
train_data_fill = train_data_raw.groupby('ptn')[fillcol].transform(lambda v: v.ffill().bfill())
```

```python
# due to pandas shenanigans, transforming the vital and test values drops the other values so we need to add them back manually
train_small_fill = pd.concat([train_small_fill,c], axis = 1)
train_data_fill = pd.concat([train_data_fill, ca],axis = 1)
train_data_fill.to_csv('./Data/train_fill.csv', na_rep='NA')
train_small_fill.to_csv('./Data/train_small_out.csv',na_rep='NA')
```

In the previous correlation section, we noted that **Bilirubin_direct** was strongly correlated to **Bilirubin_total**, which luckily has a much better observation rate after filling. That leaves just **TroponinI** and **Fibrinogen**. 

Additional research shows that Troponin is a protein found in the heart and isn't normally in the bloodstream. So the only time you should see it there is if the heart muscles are damaged. 

Fibrinogen is actually originally a chemical thrombin which is converted to fibrin during tissue or vascular injury.

It seems that both of them do sometimes have elevated values in patients with severe sepsis, but it tends to be because sepsis puts people more at risk of these types of injuries.

In other words, it seems the cause and effect are reversed. Troponin and Fibrinogen don't necessarily predict sepsis, but patients with severe sepsis seem to potentially be more at risk of complications related to elevated levels. (like how someone with AIDS is more susceptible to other common diseases, but that doesn't mean those diseases are an indicator that people have AIDS).

Because they're linked with specific injuries, there's two assumptions that seem reasonable. 1) Most of the patients tested (including those with sepsis) didn't suffer from heart or tissue/vascular injuries or perforations. 2) The tests that measure these are likely specialized (otherwise the generic bloodworks that gave us all the other values would have also provided these values and we wouldn't have so many null values still)

(this is where talking to a physician to ask which tests include these values and when they would be ordered would help a lot)

(unrelated note: Ideally we'd want to drop Unit I and Unit II as well since it a real world situation that isn't going to be a marker if you had data from 100 different ICUs but the fact is Unit I had a lot more sepsis patients than unit II)

### Feature Selection Split
At this point, we have a few options. Quite a few filter or wrapper based feature selectors can't handle null values while some others can. We have two options

1) Values that are still null after forward and backward filling means that either the patient never got that value measured or the tests they took didn't include that feature. Imputing with the mean still doesn't make sense and we don't want to throw away those null values (since they often indicate patients who didn't need test taken aka those who were likely nonsepsis) our best bet is likely to replace them with a constant that is outside the valid range of values of all features (like -999)

2) Take the current list of features we have, potentially reduce based off of things like observation rate and the preliminary SOFA features we took during our null value analysis, and run as is with a tree or gradiant boosting algorithm

https://machinelearningmastery.com/feature-importance-and-feature-selection-with-xgboost-in-python/

### Filling in with a constant
For the algorithms that can't deal with null values, we can fill them in with a constant outside of the range of valid values. These remaining values are after forward and backward filling, in other words these patients never had a test taken with these results. Imputing with the mean in that case creates false information and there's nothing to fill them with from real data. But due to the still (relatively) large percentage of null values in some of the variables we kept (such as Lactate or Bilirubin_total) discarding would loose too much data. 

HR, Age, Gender, HospAdmTime, and ICULOS have 100% observation rate, so we only need to check for the others

**Replacement strategy**
To try minimize the problems with normalization that come when you add out of range values, we'll replace NaN a value one std away from the minimum value, except for Unit1 and Unit2, which will be -1 (since they only have 2 possible values)

```python
NaNCols = ['O2Sat','Temp','SBP','MAP','DBP','Resp','BaseExcess','HCO3','FiO2','PaCO2','SaO2',
                'BUN','Calcium','Chloride','Creatinine','Glucose','Lactate','Magnesium','Phosphate',
                'Potassium','Bilirubin_total','Hct','PTT','WBC','Platelets','Unit1','Unit2']
remainHist = train_data_remain[NaNCols].hist(figsize=(20,20))
```
