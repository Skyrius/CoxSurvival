```python
from lifelines import KaplanMeierFitter
from lifelines import CoxPHFitter
import pandas as pd, numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import csv
```

```python
df = pd.read_csv("DashBoard_Data_V2.csv")

#add a new column category next to the age group. 
category = pd.cut(df.AGE_AT_MOD_START,bins=[0,40,50,60,70,110],labels=['Under40','40s','50s','60s','70+'])
df.insert(5,'Age_Group',category)

# Preview the first 5 lines of the loaded data 
df.shape
```

```python
df['AGE_AT_MOD_START'].hist(bins=10).set(xlabel='Age at Mod Start', ylabel='Count');
```

```python
sns.countplot(x='DM_HF_PVD_DEM', hue='MODALITY', data=df)
plt.xticks(rotation=60)
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
ax = df.groupby(['Age_Group','DM_HF_PVD_DEM'])['DM_HF_PVD_DEM'].count().unstack(0).plot.bar(title="Number for each Owner Name", figsize=(14,8))
_ = ax.set_xlabel('Owner')
_ = ax.set_ylabel('Frequency')
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
chart = sns.countplot(df['DM_HF_PVD_DEM'], palette='Set1');
plt.xticks(rotation=45, horizontalalignment='right')
```

```python
chart = sns.countplot(x='MODALITY', palette='Set1', hue ='Age_Group', data=df);
plt.xticks(rotation=45, horizontalalignment='right')
```

```python
sns.countplot(x='MODALITY', hue='DM_HF_PVD_DEM', data=df)
plt.xticks(rotation=60)
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
chart = sns.countplot(x='MODALITY', palette='Set1', data=df);
plt.xticks(rotation=45, horizontalalignment='right')
```

```python
groupBy = (df
            .groupby(['MODALITY', 'DM_HF_PVD_DEM'])
            .size()
            .unstack()
           )

plt.figure(figsize=(14,14))
g = sns.heatmap(
    groupBy,
    annot=True,
    linewidths=.5,
    square=True, # make cells square
    cbar_kws={'fraction' : 0.01}, # shrink colour bar
    cmap='OrRd', # use orange/red colour map
    fmt='g'
)
```

```python
ax = df.groupby(['DM_HF_PVD_DEM','Age_Group'])['Age_Group'].count().unstack(0).plot.bar(title="Number for each Owner Name", figsize=(14,8))
_ = ax.set_xlabel('Owner')
_ = ax.set_ylabel('Frequency')
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.swarmplot(data=df,x='DM_HF_PVD_DEM',y='AGE_AT_MOD_START',hue='MODALITY', size=3)
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xticks(rotation=75, horizontalalignment='right')
plt.figure(figsize=(30, 75))
```

```python
bplot=sns.stripplot(y='DM_HF_PVD_DEM', x='AGE_AT_MOD_START', 
                   data=df, 
                   jitter=True,
                   dodge=True,
                   size=4,
                   marker='o', 
                   alpha=0.5,
                   hue='MODALITY')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
