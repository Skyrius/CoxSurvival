## Problem Outline

Let's first establish our independent and dependent variables. Independent variables will the ones fed into the model (such as age and modality) and dependent variables will be the results the model returns, aka what we want to predict.

Independent Variables: 
- Modalities (PLANNED NO DIALYSIS, OPTIMAL START PD, NON-OPTIMAL START HD, OPTIMAL START HD)
- Comorbidities (4 variables) 
    - DF 
    - HF 
    - PVD 
    - Dementia
- Age

Dependent Variables:
- ED visits (number of ER visits)
- Hospital days (number of days spent in the hospital)
- Surgeries (number of surgeries)
- Hospice (number of days in hospice)
- SNF (number of days at a skilled nursing facility)

Additional Variables: (these are not part of the data given, but can be predicted through other algorithms)
- Life expectancy (aka survival) via Kaplan Meier (https://towardsdatascience.com/kaplan-meier-curves-c5768e349479) or cox regression

Kaplan Meier can only handle categorical data, meaning continuous data like age is not an option. Likewise it can only predict the effect of one variable on survival at a time.

Instead, we should use algorithms that can handle continuous varialbes, such as Cox regression. (https://towardsdatascience.com/kaplan-meier-mistakes-48cd9e168b09). Unlike KM which can be represented as simple curve, Cox regression instead gives you a hazard ratio. This variable doesn't have an innate value, instead it represents the increase in probability of suffering the event of interest (in this case, death: http://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/BS704_Survival/BS704_Survival6.html). But it can handle continuous variables and shows the effect of multiple variables on survival.

So we have two parts, a model (I'm using xgboost) that takes in Modality, Comorbidities, and Age and predicts ED visits, hospital days, number of surgers, hospice days, and SNF days.

After that, a separate model (Cox regression) is used for survival analysis

### Cox Regression

Two related probabilities are used to describe survival data: the survival probability and the hazard probability.

The survival probability, also known as the survivor function S(t), is the probability that an individual survives from the time origin (e.g. diagnosis of cancer) to a specified future time t.

The hazard, denoted by h(t), is the probability that an individual who is under observation at a time t has an event at that time.

The Cox Regression model can give us the conditional probability of an event, using the indpedent variables as covariates.

S(t|x) = P(T > t|x) in our case is S(alive>t | (Age, Modality, Comorbidities)

The Cox model is expressed by the hazard function denoted by h(t). Briefly, the hazard function can be interpreted as the risk of dying at time t. It can be estimated as follow:

h(t)=h0(t)×exp(b1x1+b2x2+...+bpxp)

where,

t represents the survival time

h(t) is the hazard function determined by a set of p covariates (x1,x2,...,xp)
the coefficients (b1,b2,...,bp) measure the impact (i.e., the effect size) of covariates.
the term h0 is called the baseline hazard. It corresponds to the value of the hazard if all the xi are equal to zero (the quantity exp(0) equals 1). The ‘t’ in h(t) reminds us that the hazard may vary over time.

The Cox model can be written as a multiple linear regression of the logarithm of the hazard on the variables xi, with the baseline hazard being an ‘intercept’ term that varies with time.

The quantities exp(bi) are called hazard ratios (HR). A value of bi greater than zero, or equivalently a hazard ratio greater than one, indicates that as the value of the ith covariate increases, the event hazard increases and thus the length of survival decreases.

Put another way, a hazard ratio above 1 indicates a covariate that is positively associated with the event probability, and thus negatively associated with the length of survival.

In summary,

HR = 1: No effect
HR < 1: Reduction in the hazard
HR > 1: Increase in Hazard

```python
# dropping combined comobidity and dependent variables
cox_data = df.drop(['PAT_ID','DM_HF_PVD_DEM','ED_VISIT','HOSPITAL_DAYS','SURGERIES','HOSPICE','SNF'], axis=1)
cox_data.head()
```

#### Handling Categorical Data
Almost all models require numerical data to calculate and any categorical data, such as the modalities and comorbidities, must be converted to their numerical representations. The easiest way is just to map each category to a unique number.

```python
#pandas can convert categorical into numerical with get_dummies
cox_dummy = pd.get_dummies(cox_data, drop_first=False) #false because I want to keep all modalities
# due to the limitations of how matrices invert, we need two of these to represent both
cox_dummy2 = cox_dummy.drop(['MODALITY_PLANNED NO DIALYSIS','DM_N','HF_N','PVD_N','DEMENTIA_N'], axis=1)
cox_dummy.drop(['MODALITY_NON-OPTIMAL START HD','DM_N','HF_N','PVD_N','DEMENTIA_N'], axis=1, inplace = True)
cox_dummy.head()
```

```python
from lifelines import CoxPHFitter

# Using Cox Proportional Hazards model
cph = CoxPHFitter()   ## Instantiate the class to create a cph object
cph.fit(cox_dummy, 'CVD_DAYS', event_col='DEADORALIVE')   ## Fit the data to train the model
cph.print_summary()    ## Have a look at the significance of the features
```

The Cox regression results can be interpreted as follow:

Statistical significance. The column marked “z” gives the Wald statistic value. It corresponds to the ratio of each regression coefficient to its standard error (z = coef/se(coef)). The wald statistic evaluates, whether the beta (β) coefficient of a given variable is statistically significantly different from 0. 

The regression coefficients. The second feature to note in the Cox model results is the the sign of the regression coefficients (coef). A positive sign means that the hazard (risk of death) is higher, and thus the prognosis worse, for subjects with higher values of that variable. 

Hazard ratios. The exponentiated coefficients , also known as hazard ratios, give the effect size of covariates. For example, exp(0.69) = 1.99, meaning Planned No Dialysis increases th baseline hazard by about 99%. So for example, if the baseline hazard at a given time was 16% chance to experience an event (death), no dialysis would increase that by 0.16 * 0.99 = 0.158 to a total of 0.318

Confidence intervals of the hazard ratios. The summary output also gives upper and lower 95% confidence intervals for the hazard ratio (exp(coef))

Global statistical significance of the model. Finally, the output gives p-values for three alternative tests for overall significance of the model: The likelihood-ratio test, Wald test, and score logrank statistics. These three methods are asymptotically equivalent. For large enough N, they will give similar results. For small N, they may differ somewhat. The Likelihood ratio test has better behavior for small sample sizes, so it is generally preferred.

From both encodings, we can see that how you change categorical data into numerical can affect your results (even though, logically, they represent the same encoding). But more importantly, we can see in both cases that Age has very little significance on survival in comparison to modality or comorbidities. When looking at multiple variables combined, Cox is the most accurate, but since Age doesn't have much of an effect, we could also look at each variable in isolation using KM

**NOTE**: For this to be accurate, we have to assume that Modality and Comorbidity are independent of each other. Which I believe is safe to assume, since Modality is what the patient (or physician) chooses as a methodology.

```python
from lifelines import KaplanMeierFitter

kmf1 = KaplanMeierFitter() ## instantiate the class to create an object
durations = cox_data['CVD_DAYS'].to_numpy()
event_observed = df['DEADORALIVE'].to_numpy()

## Two Cohorts are compared. Cohort 1. Streaming TV Not Subscribed by users, and Cohort  2. Streaming TV subscribed by the users.
groups = cox_data['MODALITY']   
i1 = (groups == 'NON-OPTIMAL START HD')      ## group i1 , having the pandas series  for the 1st cohort
i2 = (groups == 'OPTIMAL START HD')                       ## group i2 , having the pandas series  for the 2nd cohort
i3 = (groups == 'OPTIMAL START PD') 
i4 = (groups == 'PLANNED NO DIALYSIS')

## fit the model for 1st cohort
kmf1.fit(durations[i1], event_observed[i1], label='NON-OPTIMAL START HD')
a1 = kmf1.plot(ci_show=False)

## fit the model for 2nd cohort
kmf1.fit(durations[i2], event_observed[i2], label='OPTIMAL START HD')
kmf1.plot(ax=a1,ci_show=False)

## fit the model for 3st cohort
kmf1.fit(durations[i3], event_observed[i3], label='OPTIMAL START PD')
kmf1.plot(ax=a1,ci_show=False)

## fit the model for 4nd cohort
kmf1.fit(durations[i4], event_observed[i4], label='PLANNED NO DIALYSIS')
kmf1.plot(ax=a1,ci_show=False)
```
