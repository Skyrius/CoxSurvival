```python
from lifelines import KaplanMeierFitter
from lifelines import CoxPHFitter
import pandas as pd
import numpy as np
import csv
```

```python
df = pd.read_csv("KM_Data_v2.csv") 
# Preview the first 5 lines of the loaded data 
df.head()
```

```python
durations = df['CVD_DAYS'].to_numpy()
event_observed = df['DEADORALIVE'].to_numpy()
df.DM.replace(('Y', 'N'), (1, 0), inplace=True)
df.HF.replace(('Y', 'N'), (1, 0), inplace=True)
df.PVD.replace(('Y', 'N'), (1, 0), inplace=True)
df.DEMENTIA.replace(('Y', 'N'), (1, 0), inplace=True)
```

```python
## create a kmf object
kmf = KaplanMeierFitter() 

## Fit the data into the model
kmf.fit(durations, event_observed,label='Kaplan Meier Estimate')

## Create an estimate
kmf.plot(ci_show=False) ## ci_show is meant for Confidence interval, since our data set is too tiny, thus i am not showing it.
```

```python
kmf1 = KaplanMeierFitter() ## instantiate the class to create an object

## Two Cohorts are compared. Cohort 1. 
groups = df['MODALITY']   
i1 = (groups == 'NON-OPTIMAL START HD')      ## group i1 , having the pandas series  for the 1st cohort
i2 = (groups == 'OPTIMAL START HD')                       ## group i2 , having the pandas series  for the 2nd cohort
i3 = (groups == 'OPTIMAL START PD') 
i4 = (groups == 'PLANNED NO DIALYSIS')

## fit the model for 1st cohort
kmf1.fit(durations[i1], event_observed[i1], label='NON-OPTIMAL START HD')
a1 = kmf1.plot()

## fit the model for 2nd cohort
kmf1.fit(durations[i2], event_observed[i2], label='OPTIMAL START HD')
kmf1.plot(ax=a1)

## fit the model for 3st cohort
kmf1.fit(durations[i3], event_observed[i3], label='OPTIMAL START PD')
kmf1.plot(ax=a1)

## fit the model for 4nd cohort
kmf1.fit(durations[i4], event_observed[i4], label='PLANNED NO DIALYSIS')
kmf1.plot(ax=a1)
```

```python
df_noMods = df.drop(['DM_HF_PVD_DEM', 'PAT_ID', 'ED_VISIT', 'HOSP_DAYS', 'SURGERIES', 'HOSPICE', 'SNF'], axis=1)
df_noMods.head()
```

```python
df_dummy = pd.get_dummies(df_noMods)
df_dummy1 = df_dummy.drop('MODALITY_PLANNED NO DIALYSIS', axis=1)
df_dummy1.head()
```

```python
# Using Cox Proportional Hazards model
cph = CoxPHFitter()   ## Instantiate the class to create a cph object
cph.fit(df_dummy1, 'CVD_DAYS', event_col='DEADORALIVE')   ## Fit the data to train the model
cph.print_summary()    ## Have a look at the significance of the features
```

```python
cph.plot()
```

### Filter Dataset for targeted population

```python
# 0=No 1=Yes
PVD = [0,1]
DM = [0,1]
HF = [0,1]
DEMENTIA = [0,1]
MODALITY = ['OPTIMAL START HD']
    # MODALITY = ['PLANNED NO DIALYSIS', 'OPTIMAL START HD', 'OPTIMAL START PD', 'NON-OPTIMAL START HD']
PatientAge = 55
AgePlusMinus = 10

#Filter out dataframe to criteria above
iFilter = df.PVD.isin(PVD) & df.DM.isin(DM) & df.HF.isin(HF) & df.DEMENTIA.isin(DEMENTIA) & df.MODALITY.isin(MODALITY) & df.AGE_AT_MOD_START.between(PatientAge-AgePlusMinus, PatientAge+AgePlusMinus)
dfFilter = df[iFilter]

#Drop modality as its a string
dfFilter = dfFilter.drop(['MODALITY'], axis=1)

dfFilter.shape
```

### Kaplan Meier Estimate

```python
durations = dfFilter['CVD_DAYS'].to_numpy()
event_observed = dfFilter['DEADORALIVE'].to_numpy()
## create a kmf object
kmf = KaplanMeierFitter() 

## Fit the data into the model
kmf.fit(durations, event_observed,label='Kaplan Meier Estimate')

## Create an estimate
kmf.plot(ci_show=False) ## ci_show is meant for Confidence interval, since our data set is too tiny, thus i am not showing it.
```

```python
# create a custom patient
data = np.array([[1,0,1,0,85,1,0,0]])
patient = pd.DataFrame({'DM': data[:, 0], 'HF': data[:, 1], 'PVD': data[:, 2], 'DEMENTIA': data[:, 3], 'AGE_AT_MOD_START': data[:,4], 
                       'MODALITY_NON-OPTIMAL START HD': data[:5], 'MODALITY_OPTIMAL START HD': data[:6], 'MODALITY_OPTIMAL START PD': data[:,7]})
patient
```

```python
probSurvival = cph.predict_survival_function(patient)
probSurvival
```

```python
# get cumulative baseline hazards as a new pandaframe
cph.baseline_cumulative_hazard_.loc[20, 'baseline cumulative hazard']
```
